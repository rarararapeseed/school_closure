```python
import pyecharts
```

```python
from pyecharts.charts import Map,Geo
from pyecharts import options as opts
```

```python
import pandas as pd
import numpy as np
```

```python
intensity_data = pd.read_csv('cfps_intensity.csv',error_bad_lines=False,encoding='gbk')
```

```python
intensity_data
```

```python
data = []
province = intensity_data['province'].tolist()
intensity = intensity_data['intensity1'].tolist()
for i in range(31):
    data.append((province[i],intensity[i]))
```

```python
data
```

```python
map=Map()
map.add("",data,"china")
map.set_global_opts(
#     title_opts=opts.TitleOpts(title="各省撤点并校强度",subtitle="数据来源：区域经济统计年鉴",pos_right="center",pos_top="5%"),
    visualmap_opts=opts.VisualMapOpts(max_=70,min_=-6,
                                     range_color=["#D8D8D8", "#424242"]),
    ) 
map.render_notebook()
```
